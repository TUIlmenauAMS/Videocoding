<p style="line-height:1.5">
<font size="6">Linear Time Invariant Systems for time discrete signals. Again we use **bold face characters for matrices and vectors** Definition: <br><br>
    
$$\boldsymbol y( n_1, n_2 )=T(\boldsymbol x(n_1 , n_2) )$$  where T is a function, and $n_1$ and $n_2$ are integers. If T is linear, we have:<br><br>
$$T[a \cdot \boldsymbol x_1(n_1  ,n_2)+b \cdot \boldsymbol x_2(n_1 , n_2)]$$$$=a \cdot T[\boldsymbol x_1(n_1 , n_2)]+b \cdot T[\boldsymbol x_2(n_1 , n_2)]$$<br>
for real numbers a and b. This means we can “shift” the operator T before the multiplication and addition.<br>
If T is shift invariant, we have:<br><br>

$$T[\boldsymbol x(n_1 -m_1 ,n_2 -m_2)]=\boldsymbol y(n_1 -m_2 ,n_2 -m_2)$$
<br>
With $m_1$ and  $m_2$ integers.<br>
This means the operator T is the same, independent of time or place at which it is applied.<br><br>
From this we can conclude how to process signals:<br>
Any signal $\boldsymbol x(n_1,n_2)$ can be written it as a sum of 2D-unit impulse responses $\delta$, defined as<br><br>

$$\begin{equation}
 \delta(m,n)=\begin{cases}1 & \text{if m=n=0}\\
0  & \text{else}\end{cases}       
\end{equation}
$$
We take one 2D-unit pulse for each sample or pixel, with a factor which is the value of the sample or pixel at that place. we can re-write our image as a weighted sum of those unit pulses over all pixels:<br><br>

$$
\boldsymbol x ({n}_ {1} , {n} _{2} )
= \sum_{{k}_{1} = 0}^{∞} {\sum _{{k}_{2} = 0} ^ {∞} {\boldsymbol x  ({k}_{1} , {k}_{2}  ) \boldsymbol \delta  ({n} _{1} − {k} _{1} , {n}_{2} − {k}_ {2}  )}}
$$

</font></p>    


<p style="line-height:1.5">
<font size="6">
We see that for any given pixel positon $n_1$,$n_2$ the sum on the right hand side has all delta pulses euqal to zero, except the one at $k_1=n_1, k_2=n_2$, and hence we get $\boldsymbol x( n_1, n_2 )=\boldsymbol x( n_1, n_2 )$ , which shows our sum euqation is correct.<br><br>
We can now apply the rules for linear shift invariant systems to this equation:<br><br>
    
$$
T (\boldsymbol x  ({n}_ {1} , {n} _ {2}  ) ) = \sum _{{k} _{1} = 0} ^ {∞} {\sum _ {{k} _ {2} = 0} ^{∞} {\boldsymbol x  ({k}_ {1} , {k} _ {2}  ) T (\boldsymbol \delta  ({n}_ {1} − {k}_{1} , {n}_{2} − {k}_{2} ) )}}
$$    
<br>
This is our **2-dimensional convolution** equation, with<br>
$T ( \boldsymbol \delta ({n} _ {1} − {k}_ {1} , {n} _ {2} − {k} _ {2} ))$ being the 2D impulse response of our linear system T.<br><br>
This shows: all we need to determine the output of is the impulse **response**. Since we also have the shift invariance, the impulse response at location 0 is sufficient, which we call.<br> This impulse response is:<br>
$$
\boldsymbol h  ({n} _ {1} , {n} _{2}  ) ≔ T  [\boldsymbol \delta ({n}_ {1} , {n} _{2}  ) ]
$$<br><br>

**Observe:** This is the same form as in 1 dimension, but the impulse response is now 2-dimensional, hence a matrix! This is what we also called the **“filter kernel”**.<br><br>
**Observe:** Here we derived the 2-dimensional convolution. It shows that each linear time-invariant (LTI) system can be described by this convolution.<br><br>
Since we have a convolution, it makes sense to also take a look at 2-dimensional z-Transforms, because they turn a convolution into a multiplication in the z-domain.<br><br>

</font></p>

## The 2D z-Transform
        
<center></font>
<br>

<p style="line-height:1.5">
<font size="6">
    
Definition:  the 2D z-transform of x(n1,n2) is<br><br>

$$X  ({z} _{1} , {z}_{2} ) = \sum_{{n}_ {1} = 0} ^ {∞} {\sum _{{n} _ {2} = 0} ^ {∞} {\boldsymbol x ({n}_ {1} , {n}_{2} ) {z} _ {1} ^{− {n} _{1}} {{z} _{2}} ^ {− {n}_ {2}}}}$$

Observe that the 2D z-transform $X( z_1, z_2) $ is not a matrix, but a **function** of the 2 complex variables $z_1,z_2$.<br>
Example for a signal shift, signal $\boldsymbol x(n_1 -d,n_2)$:
<br><br>
$$
\sum _{{n} _{1} = 0} ^ {∞} {\sum _ {{n} _{2} = 0} ^ {∞} {\boldsymbol x (n_1 − d , {n} _ {2} ) {z} _{1} ^{− {n} _ {1}} {{z}_{2}} ^ {− {n}_{2}}}} = ...
$$<br>
Index substitution: $n {'} _{1} : = {n} _ {1} − d$:<br>
$$...=\sum_{{n '}_{1} = 0} ^ {∞} {\sum_ {{n} _ {2} = 0}^ {∞} {\boldsymbol x ({n '} _ {1} , {n} _ {2} ) {z} _ {1} ^ {− ({n '} _ {1} + d )} {{z} _ {2}} ^ {− {n} _ {2}}}} = {z} _ {1} ^{− d} ⋅ \boldsymbol X  ({z} _ {1} , {z} _ {2} )$$

We see: a **shift** in the spacial domain corresponds to a **multiplication** with ${z}_{1} ^ {− d}$ in the z-domain.<br>
**Example** for a z-transform:<br>
$$\boldsymbol x  ({n} _ {1} , {n} _ {2} ) = {a} ^ {{n} _ {1}} {b} ^ {{n} _ {2}}$$
<br>
for $n_1 , n_2 >= 0$ and a and b real or complex numbers.<br>
Its resulting z-transform is:<br><br>
$$
 X ({z} _ {1} , {z} _ {2}  ) = \sum _{{n} _ {1} = 0} ^ {∞} {\sum _ {{n} _ {2} = 0}^ {∞} {{a} ^ {{n} _ {1}} {b} ^ {{n} _ {2}} {z} _{1}^ {− {n} _{1}} {{z} _ {2}} ^ {− {n}_{2}}}}
$$

<br>
$$
 X ({z} _{1} , {z} _ {2}  ) = \sum _ {{n} _{1} = 0}^ {∞} {{ ({az} _ {1} ^{− 1}  )} ^ {{n} _{1}}} \sum _ {{n} _ {2} = 0} ^ {∞} {{ (b {{z} _ {2}} ^ {− 1}  )} ^ {{n}_{2}}}
$$
Using that this is a geometric sum, which has a closed form solution, yields the final result:

<br><br>
$$
X ({z} _ {1} , {z}_{2} ) = {\frac{1}{1 − a {z} _ {1} ^ {− 1}}} \frac{1}{1 − b {z} _{2} ^ {− 1}}
$$<br>
Observe: for $\mid{ z_1} \mid>= 1$ and $\mid{ z_2}\mid >= 1$  and for $\mid {a} \mid <1$ and $ \mid{b}\mid<1 $ the sum converges, we obtain a finite result and hence we have a stable system. For this convergence the sum is evaluated on the unit circle for the z’s.<br><br>
The condition $\mid{ z_1}\mid >\mid{a}\mid$ and $ \mid { z_2}\mid >\mid{b}\mid$, which is necessary for convergence, is also called the Region of Convergence. This is a circular region outside |a| and |b|.

<br><br>



To determine stability of a 2 dimensional system, we can also consider the location of the poles of above expression in the  complex z-domain. That means in this case we have the poles of the z-transform at the locations $z_1=a$ or $z_2=b$, which are **inside the unit circles** (we have 2 circles, one for each dimension, for and!), and we obtain a **stable system**. **Both** poles need to be inside their unit circle to obtain a stable system.<br>
This is like in the 1-dimensional case, just that we now have two z-transform planes.<br>

## 2-D Signal Processing
## Properties of the 2D z-Transform    
<center></font>
<br>

<p style="line-height:1.5">
<font size="6">
    
-Linearity:
<br><br>
$$
a \boldsymbol x  ({n} _{1} , {n} _ {2} ) + b \boldsymbol y ({n} _ {1} , {n} _ {2}  ) ↔ a  X  ({z}_ {1} , {z} _ {2}  ) + b  Y  ({z} _ {1} , {z} _ {2} )
$$<br>
-Convolution:<br><br>
$$
\boldsymbol x ({n}_ {1} , {n} _ {2} ) ∗ \boldsymbol y  ({n} _ {1} , {n} _ {2}  ) ↔X  ({z} _ {1} , {z} _ {2} )  Y  ({z} _ {1} , {z} _ {2} )
$$
<br>
-Separable Sequence:<br><br>
$$
\boldsymbol {x} _ {1} ({n} _ {1} ) \boldsymbol {x} _ {2}  ({n} _ {2}  ) ↔  {X}_{1} ({z}_{1}  )  {X}_{2}  ({z}_{2}  )
$$<br>
-Shift of a Sequence:<br><br>
$$
\boldsymbol x  ({n} _{1} − {m} _ {1} , {n} _{2} − {m} _ {2}  ) ↔  X ({z} _{1} , {z} _{2}  ) {z} _{1} ^ {{− m} _{1}} {{z} _ {2}} ^ {− {m} _ {2}}
$$
<br>
-Symmetry Property:<br><br>
$$
\boldsymbol x  (− {n} _ {1} , − {n} _ {2} ) ↔  X  ({z} _{1} ^ {− 1} , {z} _ {2} ^{− 1} )
$$<br>
-Inverse z-Transform:<br><br>
$$
\boldsymbol x  ({n} _ {1} , {n} _ {2}  ) = \frac{1}  {{(2πj  )} ^ {2}} \oint_{{C} _ {1} }\oint_{{C} _ {2}} { X  ({z}_  {1} , {z} _{2}  ) {z} _{1} ^ {{n} _ {1} − 1}} {z} _{2} ^ {{n} _ {2} − 1} d {z} _{1} d {z} _{2}
$$
<br>
and $C_1$ are $C_2$ curves that include all poles of $X(z_1,z_2)$. If the system is stable, all poles are within the unit circle, and $C_1$ and $C_2$ can be unit circles for dimension 1 and 2. This is then a special case which corresponds to the inverse Discrete Time Fourier Transform.<br>
More practical: write z-transform as sum of low order polynomial fractions:
<br><br>
$$
X  (z ) = \frac{N  (z  )} {D (z)} = {N} _{0}  (z ) + \frac{{N} _ {1}  (z  )}  {{D} _ {1} (z) {D} _ {2} (z ) ⋯ {D}_{M}  (z )}
$$
<br>
$$
 = {N} _ {0}  (z  ) + \frac {{N}_{2} (z  )}  {{D} _{1}  (z  )} + \frac{{N} _ {3}  (z )} {{D} _ {2}  (z  )} + ⋯ + \frac{{N} _ {M + 1}  (z  )}{{D} _ {M} (z )}
$$<br><br>

And then inverse transform by inspection of the fraction terms. (From: J.S. Lim, Two Dimensional Signal and Image Processing). Each of the denominators usually corresponds to a pole of the z-transform. Hence we get a sum of fractions with simple poles in the denominator. For simple poles we know the corresponding function in the time or space domain.<br><br>

**Example:**
If we observe the following z-transform (as one of the summands above),<br>
$$
X  ({z}_ {1} , {z} _ {2} ) = \frac{1} {1 − a {z} _{1} ^{− 1}} \frac{1} {1 − b {z} _{2}^ {− 1}}
$$
<br>
we know that the space domain signal is.<br><br>
$$x  ({n} _{1} , {n} _ {2}  ) = {a} ^ {{n} _ {1}} {b} ^ {{n}_ {2}}
$$
This can be done for all summands and then the result added up accordingly. So if we know the poles in the z-domain, we know the exponetial terms in the spacial domain.<br>

</font></p>    

<p style="line-height:1.5">
<font size="6">
    
<br>    
**Example:** After partial fraction decomposition of our 2D z-transform we obtain<br><br>

$$
X  ({z}_{1} , {z}_{2}  ) = \frac{1}{1 − 0.5 {z} _{1} ^ {− 1}} \frac{1}  {1 − 0.3 {z} _ {2} ^ {− 1}} + \frac{1} {1 − 0.7 {z} _ {1} ^ {− 1}} \frac{1}  {1 − 0.4 {z} _ {2} ^ {− 1}}
$$
<br>

with poles at position $z_1=0.5$ and $0.7$ and $z_2=0.3$ and $0.4$.<br>
The inverse z-transform is then<br>
$$
x ({n} _ {1} , {n} _ {2}  ) = {0.5} ^ {{n} _ {1}} ⋅ {0.3} ^ {{n} _ {2}} + {0.7} ^ {{n} _ {1}} ⋅ {0.4} ^ {{n} _ {2}}
$$

# 2D Filter Analysis, Examples
        
<center></font>
<br>

<p style="line-height:1.5">
<font size="6">
We saw that down- and up-sampling creates spectral copies of our signal and **aliasing artifacts**, and that we can **avoid them using a low pass filter**. This low pass filter supresses the high frequencies, which appear towards the middle of our 2D DFT of the image, and keep the low frequencies, which appear near the corners of the 2D DFT. A **convolution** of our image with a filter kernel corresponds to a **multiplication** of the 2D DFT of the image with the 2D DFT of the filter kernel. Hence ideally the 2D DFT of our filter kernel (its 2D frequency response) should have **values 1 near the corners and zeros towards the center.**<br>
To see if this is the case, we take our rectangular filter kernel and analyse it:<br><br>
    
</font></p>  

```python
import numpy as np
h1=np.ones((8,8))
print(h1)
```


<p style="line-height:1.5">
<font size="6">
We now compute the 2D-frequency response of the rectangular filter kernel:
</font></p>    

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
h1=np.ones((8,8))
A=np.abs(np.fft.fft2(h1,(32,32)));

[r,c]=A.shape
x=range(c)
y=range(r)
X,Y=np.meshgrid(x,y)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X,Y,A)
ax.set_xlabel('k_2')
ax.set_ylabel('k_1')
ax.set_zlabel('Value')
plt.title('2D frequency response of \
          rectangular filter kernel with size 32 FFT')

plt.show()
```


<p style="line-height:1.5">
<font size="6">
Now we analyse our pyramidal filter kernel:

</font></p>    

```python

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import scipy.signal
import numpy as np
h1=np.ones((8,8))
h2=scipy.signal.convolve2d(h1,h1)/8
A=np.abs(np.fft.fft2(h2,(32,32)));
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X,Y,A)
ax.set_xlabel('k_2')
ax.set_ylabel('k_1')
ax.set_zlabel('Value')
plt.title('2D frequency response of pyramidal filter\
kernel with size 32 FFT')
plt.show()
```


<p style="line-height:1.5">
<font size="6">Compare it with the frequency response of our rectangular filter kernel. We see that the side lobes are much smaller, which means the pyramidal filter kernel suppresses the higher frequencies better, and hence it is a **better low pass filter**, which better suppresses aliasing artifacts in sampling.

</font></p>
